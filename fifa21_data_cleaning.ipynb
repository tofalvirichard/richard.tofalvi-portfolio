```python
data["Height"] = data["Height"].str.replace("cm", "", regex=False)
data["Height"] = data["Height"].str.replace('"', '', regex=False)
for i in range(len(data)):
    value = data.loc[i,"Height"]
    if "'" in  value :
        parts = value.split("'")
        value_cm = float(parts[0]) * 30.48 + float(parts[1])* 2.54
        data.loc[i,"Height"] = round(value_cm,0)
data["Height"]= data["Height"].astype(int)       
# Convert foot and inch to cm, remove cm + convert series to integer
```

```python
data["Weight"].unique()
```

```python
data["Weight"] = data["Weight"].str.replace("kg", "", regex=False)
mask = data["Weight"].str.contains("lbs", na=False)
data.loc[mask, "Weight"]=round(data.loc[mask, "Weight"].str.replace("lbs","", regex=False).astype(int) / 2.205,0)
data["Weight"]= data["Weight"].astype(int)
#lbs to cm and remove cm + plus convert series to integer
```

```python
for col in data.select_dtypes(include="object").columns:
    data[col] = data[col].str.replace(r"\s+", " ", regex=True).str.strip()
data
# remove all whitescpace, included leading/trailing whitespace
```

```python
data["Joined"]=pd.to_datetime(data["Joined"], errors="coerce")
data["Joined"].isnull()
invalid_dates = data[data["Joined"].isna()]
print(invalid_dates)
# I checked that Joined column contains only dates
```

```python
from datetime import datetime
today = pd.Timestamp.today()
data["years_since_joined"] = (today - data["Joined"]).dt.days / 365.25
players_10_years = data[data["years_since_joined"] >= 10]
print(players_10_years[["Name", "Club", "Joined", "years_since_joined"]])
# I listed players who's played more than 10 years in same club
```

```python
data["Value"] = data["Value"].str.replace("€","",regex=False)
mask_M = data["Value"].str.contains("M",regex=False)
mask_K = data["Value"].str.contains("K",regex=False)
data["Value"] = data["Value"].str.replace("K","",regex=False)
data["Value"] = data["Value"].str.replace("M","",regex=False)
data.loc[mask_K,"Value"] = data.loc[mask_K,"Value"].astype(float) * 1000
data.loc[mask_M,"Value"] = data.loc[mask_M,"Value"].astype(float) * 1000000
#transform to number
```

```python
data["Wage"].unique()
```

```python
data["Wage"] = data["Wage"].str.replace("€","",regex=False)
mask_M = data["Wage"].str.contains("M",regex=False)
mask_K = data["Wage"].str.contains("K",regex=False)
data["Wage"] = data["Wage"].str.replace("K","",regex=False)
data["Wage"] = data["Wage"].str.replace("M","",regex=False)
data.loc[mask_K,"Wage"] = data.loc[mask_K,"Wage"].astype(float) * 1000
data.loc[mask_M,"Wage"] = data.loc[mask_M,"Wage"].astype(float) * 1000000
#transform to number
```

```python
data["Wage"].unique()
```

```python
cols_with_star = [col for col in data.columns if data[col].astype(str).str.contains("★").any()]

print(cols_with_star)
for serias in cols_with_star:
    data[serias] = data[serias].str.replace("★","",regex=False).astype(int)
```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

x = data["Wage"]
y = data["Value"]

def millions_formatter(x, pos):
    if x >= 1e6:
        return f"{x*1e-6:.0f}M"
    elif x >= 1e3:
        return f"{x*1e-3:.0f}K"
    else:
        return str(int(x))

plt.figure(figsize=(8,6))
plt.scatter(x, y, alpha=0.5)
plt.xlabel("Wage (€)")
plt.ylabel("Value (€)")
plt.title("Játékosok Value vs Wage")
plt.grid(True)

# Y-tengely formázása
plt.gca().yaxis.set_major_formatter(FuncFormatter(millions_formatter))
plt.gca().xaxis.set_major_formatter(FuncFormatter(millions_formatter))

plt.show()
```
